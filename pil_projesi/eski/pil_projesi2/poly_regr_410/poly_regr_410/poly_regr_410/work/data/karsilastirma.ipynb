```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data correctly
df = pd.read_csv("data/karsilastirma.csv", delimiter=';')

# Remove the " V" and quotes
df = df.replace('"', '', regex=True)
df = df.replace(' V', '', regex=True)

# Now convert needed columns to float
for col in df.columns[1:6]:  # assuming 2nd to 6th columns are your voltage data
    df[col] = df[col].astype(float)

# Prepare data for plotting
X = np.arange(len(df))
y1 = df.iloc[:,1]
y2 = df.iloc[:,2]
y3 = df.iloc[:,3]
y4 = df.iloc[:,4]
y5 = df.iloc[:,5]

# Plotting
a = 104   # beginning of prediction number
fig, ax = plt.subplots(1,1,figsize=(12,6))
plt.plot(X + a, y1, "--", color = "orangered", label="ARIMA")
plt.plot(X + a, y2, "--", color = "green", label="Poly. Regr.")
plt.plot(X + a, y3, "--", color = "black", label="Lin. Regr.", lw=1.5)
plt.plot(X + a, y4, "--", color = "purple", label="RANSAC")
plt.plot(X + a, y5, "-+", color = "blue", label="Grad. Desc.", lw=0.5)
plt.title("Comparison", fontsize = 14)
plt.xlabel("Prediction Nr.")
plt.ylabel("Voltage Level [V]")

plt.legend(loc='upper left')
plt.grid()
plt.show()
```
